```python
"""
Побудуйте класифікатор на основі дерева ухвалення рішень, використовуючи датасет ірісів Фішера - тренувальний набір 
ознак по трьох видах ірісів.
"""
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import graphviz
#from sklearn.tree import export_graphviz
import subprocess
from sklearn import tree
```

```python
# Завантаження датасету ірисів Фішера
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Розподіл даних на тренувальний та тестовий набори
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Створення та навчання класифікатора на основі дерева прийняття рішень
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
# Застосування навченого класифікатора до тестових даних
y_pred = classifier.predict(X_test)
```

```python
# Оцінка точності класифікатора
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Генерація файлу dot для візуалізації дерева
dot_data = export_graphviz(classifier, out_file=None,
                           feature_names=iris.feature_names,
                           class_names=iris.target_names,
                           filled=True, rounded=True,
                           special_characters=True)
graph = graphviz.Source(dot_data)
```

```python
# Сохранение файла dot
dot_file = "decision_tree.dot"
with open(dot_file, "w") as f:
    f.write(dot_data)
```
